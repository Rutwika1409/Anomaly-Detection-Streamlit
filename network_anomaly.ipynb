```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import IsolationForest
```

# Step 1: Load dataset

```python
data = pd.read_csv('embedded_network_security.csv')
data.head()
```

```python
data.isna().sum()
```

```python
# check information about the features
data.info()
```

## Step 2: Data Preprocessing


```python
features = data.drop(columns = ['label'])
features.head()
```

```python
for col in features.columns:
    if features[col].dtype == 'bool':
        features[col] = features[col].astype(int)
```

```python
features.head()
```

```python
features.info()
```

## Step 3: Feature Engineering

```python
# scale all the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_features
```

```python
# Convert the scaled features to a dataframe
# Columns = features.column given so that column names don't chnage in dataframe
# Can customize the column names, if none given column names will be numbers
scaled_features_df = pd.DataFrame(scaled_features, columns = features.columns) 
scaled_features_df.head()
```

```python
# Parameters for isolation forest

n_estimators = 100
contamination = 0.04
sample_size = 256
```

```python
# Train the isolation forest model
model = IsolationForest(n_estimators = n_estimators, contamination = contamination, max_samples = sample_size, random_state = 42)
model.fit(scaled_features_df)

# Predict anomalies
anomaly_labels = model.predict(scaled_features_df)
anomaly_labels
```

```python
scaled_features_df['anomaly'] = anomaly_labels
```

```python
# Count the number of anomalies and normal data points
anomaly_counts = scaled_features_df['anomaly'].value_counts()
print(anomaly_counts)
```

```python
# Visualization anomalies using scatter plot
plt.figure(figsize=(10,6))

# Plot normal instances
normal_data = scaled_features_df[scaled_features_df['anomaly']==1]
# Here index is used to get the row number and 'anomaly' column is used to get the value of anomaly which is 1 for normal data points
# Here alpha is used to make the points transparent so that we can see the overlapping points clearly
plt.scatter(normal_data.index, normal_data['anomaly'], color='blue', label='Normal', alpha=0.5)

# Plot anomalies
anomalies = scaled_features_df[scaled_features_df['anomaly']==-1]
plt.scatter(anomalies.index, anomalies['anomaly'], color='red', label='Anomaly', alpha=0.5)
plt.xlabel('Index')
plt.ylabel('Anomaly')
plt.title('Anomaly Detection using Isolation Forest')
plt.show()
```

```python
# Visualize anomalies - Multiple feature comparisons
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
 
normal = scaled_features_df[scaled_features_df['anomaly'] == 1]
anomaly = scaled_features_df[scaled_features_df['anomaly'] == -1]
 
# Plot 1: CO vs C6H6
axes[0].scatter(normal['src_port'], normal['dst_port'], color='blue', label='Normal', alpha=0.6, s=30)
axes[0].scatter(anomaly['src_port'], anomaly['dst_port'], color='red', label='Anomaly', alpha=0.8, s=50, marker='x')
axes[0].set_xlabel('src_port')
axes[0].set_ylabel('dst_port')
axes[0].set_title('src_port vs dst_port')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
 
# Plot 2: NOx vs NO2
axes[1].scatter(normal['packet_size'], normal['inter_arrival_time'], color='blue', label='Normal', alpha=0.6, s=30)
axes[1].scatter(anomaly['packet_size'], anomaly['inter_arrival_time'], color='red', label='Anomaly', alpha=0.8, s=50, marker='x')
axes[1].set_xlabel('packet_size')
axes[1].set_ylabel('inter_arrival_time')
axes[1].set_title('packet_size vs inter_arrival_time')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
 
plt.tight_layout()
plt.show()
 
print(f"Normal data points: {len(normal)}")
print(f"Anomalies detected: {len(anomaly)}")
 
```
